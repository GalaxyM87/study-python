\begin{equation}
\sigma=\sigma_{T} \cdot \frac{3}{4}\left[\frac{1+x}{x^{3}}\left\{\frac{2 x(1+x)}{1+2 x}-\ln (1+2 x)\right\}+\frac{1}{2 x} \ln (1+2 x)-\frac{1+3 x}{(1+2 x)^{2}}\right]
\end{equation}

\begin{equation}
\varepsilon_{f}=\frac{\varepsilon_{i}}{1+\frac{\varepsilon_{i}}{m_{e} c^{2}}(1-\cos \theta)}
\end{equation}

\begin{equation}
\frac{d \sigma}{d \Omega}=\frac{r_{0}^{2}}{2} \frac{\varepsilon_{f}^{2}}{\varepsilon_{i}^{2}}\left(\frac{\varepsilon_{i}}{\varepsilon_{f}}+\frac{\varepsilon_{f}}{\varepsilon_{i}}-\sin ^{2} \theta\right)
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def final(xi,theta):
    core1 = xi / (1 + xi * (1 - np.cos(theta)))
    return core1

def sigma(xi,theta):
    xf = final(xi,theta)
    core2 = ((xf**2) / (xi**2)) * ( xi/xf + xf/xi - (np.sin(theta))**2 )
    return core2

x = np.array([0.00001, 0.5, 10, 100, 1000])

theta = np.arange(0,360,0.1)
theta = theta / 180 * np.pi

fig = plt.figure(figsize=(8,8))
ax = plt.subplot(111, projection='polar')

for xi in x:
    s = sigma(xi,theta)
    ax.plot(theta,s,label = f"x={xi}")
    
ax.legend()
fig.savefig("2.png",dpi = 600)
```
